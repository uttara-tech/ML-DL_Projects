```python
sns.countplot(x='gender',data=data,hue='gender')
plt.ylabel('frequence')
plt.title('Gender Bar Plot')
plt.show()
```

```python
sns.countplot(x='parental level of education',data=data,hue='parental level of education')
plt.xticks(rotation=45)
plt.ylabel('frequence')
plt.title('Parental level of education Bar Plot')
plt.show()
```

```python
sns.barplot(x='gender',y='reading score',data=data,hue='gender')
plt.show()
```

```python
sns.barplot(x='gender',y='reading score',hue='race/ethnicity',data=data)
plt.show()
```

```python
sns.histplot(data['math score'],bins =10,kde=True)
```

```python
sns.jointplot(x='math score',y='writing score',data=data)
```

```python
sns.pairplot(data)
```

```python
sns.boxplot(x='gender',y='math score',data=data,hue='gender')
plt.show()
```

```python
sns.boxplot(x='gender',y='writing score',data=data,hue='gender')
plt.show()
```

```python
sns.boxplot(x='race/ethnicity',y='writing score',data=data,hue='race/ethnicity')
plt.show()
```
